```python
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import Math,display
from sympy.abc import *
```

```python
expr = (1/a)
```

```python
display(Math('f(x) = '+sym.latex(expr)))
```

```python
for x in range(1,10):
    y = expr.subs({a:x})
    plt.plot(x,y,'o', label='f(%g) = %g'%(x,y))
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x) = %s'%(expr))
plt.show()
```

```python
x = sym.sqrt(abs(z))
display(Math(sym.latex(x)))
```

```python
for i in range(-20,21):
    y = x.subs({z:i})
    plt.plot([0,i],[0,y])
plt.grid()
plt.show()
```
