```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
from IPython.display import Math,display
from sympy.abc import *
```

```python
expr = (1/a)
```

```python
display(Math('f(x) = '+sym.latex(expr)))
```

```python
for x in range(1,10):
    y = expr.subs({a:x})
    plt.plot(x,y,'o', label='f(%g) = %g'%(x,y))
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x) = %s'%(expr))
plt.show()
```

```python
x = sym.sqrt(abs(z))
display(Math(sym.latex(x)))
```

```python
for i in range(-20,21):
    y = x.subs({z:i})
    plt.plot([0,i],[0,y])
plt.grid()
plt.show()
```

```python
plt.plot([0,0],[0,2])
plt.plot([0,2],[0,0])
plt.plot([0,2],[2,2])
plt.plot([2,2],[2,0])
plt.axis([-8,8,-8,8])
plt.show()
```

```python
expr = m*x + b
for i in range(-20,21):
    plt.plot(i,expr.subs({x:i, m:2, b:2}),'o')
plt.show()
```

```python
x = [-20,20]
m = 3
b = 7
y = list(map(lambda i: (m*i + b) ,x))
plt.plot(x,y)
plt.xlim(x)
plt.ylim(x)
axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'--k')
plt.plot([0,0],axis.get_ylim(),'--k')
plt.grid()
plt.show()
```

```python
x = [-20,20]
m = [-5,5]
b = [2.1,7]
for i in range(len(x)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = np.linspace(-4,4,200)
for i in range(-1,4):
    y = x ** i
    plt.plot(x,y,label='$y = x^{%g}$'%(i))
plt.xlim([-4,4])
plt.ylim([-20,20])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
